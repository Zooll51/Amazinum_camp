```python
import pandas as pd
import numpy as np
```

```python
df_train = pd.read_csv(r"titanic\train.csv")
df_test = pd.read_csv(r"titanic\test.csv")
df_train
```

```python
df_test
```

```python

df_train.isnull().sum()
```

```python
def remake_age(x):
    if x <= 12:
        return "child"
    elif x >= 13 and x <= 18:
        return "teen"
    elif x >= 19 and x <= 45:
        return "young"
    elif x >= 46 and x <= 65:
        return "mid"
    elif x >= 66:
        return "old"
    else:
        return np.nan
df_train["Age"] = df_train["Age"].apply(remake_age)
df_test["Age"] = df_test["Age"].apply(remake_age)
df_train
```

```python
df_train['Age'].value_counts()
```

```python
df_train["Age"] = df_train["Age"].fillna("young")
df_train['Age'].value_counts()
```

```python
df_train.isnull().sum()
```

```python
gender_sibmission = pd.read_csv(r"titanic\gender_submission.csv", index_col="PassengerId")
gender_sibmission
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

```python
features = ["Pclass", "Sex", "Age", "SibSp", "Parch"]
X = pd.get_dummies(df_train[features])
y = df_train["Survived"]
X_test = pd.get_dummies(df_test[features])
X
```

```python
clf = SVC(kernel="rbf")
grid_values = {'gamma': [0.001, 0.01, 0.05, 0.1, 1, 10, 100]}

grid_clf = GridSearchCV(clf, param_grid=grid_values, return_train_score=True).fit(X, y)
clf_score = grid_clf.best_score_
print('Grid best parameter (max. accuracy): ', grid_clf.best_params_)
print('Grid best score (accuracy): ', grid_clf.best_score_)
```

```python
grid_clf_auc = GridSearchCV(clf, param_grid = grid_values, scoring = 'roc_auc').fit(X,y)
print('Grid best parameter (max. AUC): ', grid_clf_auc.best_params_)
print('Grid best score (AUC): ', grid_clf_auc.best_score_)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf_forest = RandomForestClassifier(random_state=0)
params_grid = {'max_depth': np.arange(3, 10),
              'max_leaf_nodes':list(range(20,100,20)),
              'max_features':list(range(2,10,2))
               }
grid_clf = GridSearchCV(clf_forest, param_grid=params_grid).fit(X,y)
print('Grid best parameter: ', grid_clf.best_params_)
print('Grid best score: ', grid_clf.best_score_)
```

```python
clf = RandomForestClassifier(max_depth=3, max_features=2, max_leaf_nodes=20).fit(X,y)
pred = clf.predict(X_test)
res = pd.DataFrame({'PassengerId': df_test.PassengerId, 'Survived': pred})
res.to_csv('submission_rand_forest_with_age.csv', index=False)
```
