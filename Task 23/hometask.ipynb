```python
import numpy as np
import pandas as pd

import nltk
from nltk.corpus import stopwords
stop_words= set(stopwords.words('english'))
```

```python
fn='Archive/rt-polarity.neg'
content_neg = ''
with open(fn, "r",encoding='utf-8', errors='ignore') as f: # some invalid symbols encountered 
    content_neg = f.read()  
texts_neg=  content_neg.splitlines()
print ('len of texts_neg = {:,}'.format (len(texts_neg)))
for review in texts_neg[:5]:
    print ( '\n', review)
```

```python
fn='Archive/rt-polarity.pos'
content_pos = ''

with open(fn, "r",encoding='utf-8', errors='ignore') as f:
    content_pos = f.read()
texts_pos=  content_pos.splitlines()
print ('len of texts_pos = {:,}'.format (len(texts_pos)))
for review in texts_pos[:5]:
    print ('\n', review)
```

```python
def preprocess(text):
    return nltk.tokenize.word_tokenize(text.lower())

all_content = content_pos + '' + content_neg
all_words = preprocess(all_content)
print (len(all_words))
print(all_words[:100])
```

```python
all_words=nltk.FreqDist(all_words)
print ('len of vocabulary: {:,}'.format (len(all_words)))
# Use most common words
most_common_words = list(zip(*all_words.most_common()))[0] # [0] means names whereas [1] are frequencies 
print (most_common_words[:100])
```

```python
def remove_stop_words(words):  
    return [w for w in words if w not in stop_words]

most_common_words_filtered = remove_stop_words(most_common_words)
```

```python
word_features = most_common_words_filtered [:3000]
print (word_features[:100])
```

```python
def my_tokenizer(s):
    s = s.lower() # downcase
    tokens = nltk.tokenize.word_tokenize(s) # split string into words (tokens)
    tokens = [t for t in tokens if t not in stop_words] # remove stopwords
    return tokens

positive_tokenized = []
negative_tokenized = []

for review in texts_pos:
    tokens = my_tokenizer(review)
    positive_tokenized.append(tokens)

for review in texts_neg:
    tokens = my_tokenizer(review)
    negative_tokenized.append(tokens)

category = ['neg', 'pos']

reviews_neg = [(tokens, category[0]) for tokens in negative_tokenized]
reviews_pos = [(tokens, category[1]) for tokens in positive_tokenized]

all_reviews = reviews_neg + reviews_pos
np.random.shuffle(all_reviews)
print(all_reviews[:10])
```

```python
def find_features(review_tokens):
    return {w: w in set(review_tokens) for w in word_features}

data_set= [(find_features(review_tokens), category) for (review_tokens, category) in all_reviews]
```

```python
data_set[0]
```

```python
split_on = int(len(data_set)*.8)
X_y_train= data_set[:split_on]
X_y_test = data_set[split_on:]
print (len(X_y_train))
```

```python
clf= nltk.NaiveBayesClassifier.train(X_y_train)
nltk.classify.accuracy(clf, X_y_test)*100
```
